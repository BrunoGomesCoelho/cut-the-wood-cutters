```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 50)

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("sample_data_short.csv", sep=";")
df.head()
```

```python
graph_df = df.groupby(["numero_serie_dof", "especie_agrupada", "produto_agrupado"]).first()
graph_df.drop(columns=["volume", "valor"], inplace=True)
```

```python
graph_df[["volume", "valor"]] = df.groupby(["numero_serie_dof", 
                                                   "especie_agrupada", 
                                                   "produto_agrupado"])["volume", "valor"].sum()
```

```python
idx = ~graph_df["cpf_remetente"].str.contains("\*")
temp = graph_df["nome_remetente"].copy()
temp.loc[idx] = ""
graph_df["src_id"] = graph_df["cpf_remetente"].str.replace("\*", "")+ temp

idx = ~graph_df["cpf_destinatario"].str.contains("\*")
temp = graph_df["nome_destinatario"].copy()
temp.loc[idx] = ""
graph_df["dst_id"] = graph_df["cpf_destinatario"].str.replace("\*", "") + temp
```

```python
graph_df["src_id"] = graph_df["cpf_remetente"] + temp
```
